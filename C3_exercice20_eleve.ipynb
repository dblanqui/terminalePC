            <li>La concentration $[H_3O^+]$ en ions oxoniums est donc : $\large [H_3O^+]=\frac{C_A×V_A– C_B× V_0}{V_0+ V_A}$ <b>(ligne 25 du script à compléter)</b></li><br/>
        <li>Le pH est donné par la relation $\large pH=- log \frac{[H_3O^+]}{c°} = -log \frac{C_A × V_A– C_B× V_0}{V_0 + V_A}$ (ligne 26 du script)</li>

```python
from math import log10
### Bibliothèques importées
import numpy as np
import matplotlib.pyplot as plt

### Définition des variables
'''CB, CBH, CH : liste des concentrations respectives des espèces B–,BH, H3O+, du pH et du volume versé à l'équivalence'''
CBi,CA,V0,Ve = 0.1, 0.1, 10, 10

### Initialisation de 5 listes vides (concentrations en B-, BH, H+ et pH et volume
CB, CBH, CH, pH, V =[], [], [], [], []

### Remplissage des 5 listes suivant le volume VA versé : append ajoute la valeur entre parenthèses en fin de liste
for VA in np.arange(1,26,0.5) : #on répete les opérations pour chaque 0,5 mL de 1 à 26 mL
    if VA < Ve : #Avant l'équivalence
        V.append(VA)
        CB.append((CBi*V0-CA*VA)/(V0+VA))
        CBH.append((CA*VA/(V0+VA)))
        CH.append(0)
        pH.append(4.2 + log10((0.1*V0-CA*VA)/(CA*VA))) #le pH est donné par la relation pH=4,2 + log([B-]/[BH])
    elif VA > Ve:
        V.append(VA)
        CB.append(0)
        CBH.  # A compléter
        CH. # A compléter
        pH.append(-log10((CA*VA-CBi*V0)/(V0+VA))) #le pH est donné par la relation pH=-log([H3O+])

### Affichage des graphiques
plt.figure(figsize=(16,10))
plt.subplot(121)
plt.plot(V, CB, 'bx-', label="CB")
plt.plot(V, CBH, 'rx-', label="CBH")
plt.legend()
plt.subplot(122)
plt.ylim(0,14)
plt.plot(V, pH, 'rx-', label="pH")
plt.legend()
plt.show()
```
